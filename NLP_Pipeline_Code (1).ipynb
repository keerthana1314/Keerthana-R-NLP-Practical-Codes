```python
import nltk
from nltk.tokenize import word_tokenize

nltk.download('averaged_perceptron_tagger_eng') # Download the required resource
nltk.download('maxent_ne_chunker_tab') # Download the required resource for NE chunking

# Assuming 'text' is globally available from a previous cell.
# If not, it would need to be defined here as well, e.g., text = "...".
# Based on the notebook state, 'text' is available.

tokens = word_tokenize(text) # Tokenize the text to get 'tokens'
pos_tags = nltk.pos_tag(tokens) # Perform POS tagging to get 'pos_tags'

named_entities = nltk.ne_chunk(pos_tags)
print("Named Entities:", named_entities)
```
